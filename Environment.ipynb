```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#imporing the csv file to be analyzed

df=pd.read_csv('environmental-tax-account-1999-2020-csv.csv')
df
```

```python
#This code is showing the sum of the missing values in the dataset

missing=df.isnull().sum()
print(missing)
```

```python
#Showing us the full table and columns with the missing values

df.dropna(how='all')
```

```python
#This code is to help drop the columns with the missing values

newdata=df.dropna(axis = 1)
newdata
```
