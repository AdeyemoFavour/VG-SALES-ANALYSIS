```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#IMPORTING THE CSV FIE TO BE ANALYSED

props=pd.read_csv('property data.csv')
props
```

```python
#DATATYPES OF EACH COLUMN

result=props.dtypes
result
```

```python
#CONVERSION OF COLUMN SQ_FT FROM OBJECT  DATA TYPE TO FLOAT DATA TYPE TO AID REPLACEMENT

props['SQ_FT'] = pd.to_numeric(props['SQ_FT'],errors = 'coerce')
props['SQ_FT']
```

```python
#NEW VALUES OF COLUMN SQ_FT AFTER REPLACEMENT

mean=props["SQ_FT"].mean()
props["SQ_FT"].replace(np.nan,mean)
```

```python
#FILLING THE REPLACED  VALUES OF COLUMN SQ_FT INTO THE DATASET

props["SQ_FT"] = props["SQ_FT"].fillna(props["SQ_FT"].mean())
```

```python
#SHOWING THAT THE MISSING VALUES OF COLUMN SQ_FT HAS BEEN FILLED UP

props.head(9)
```

```python
#NEW VALUES OF COLUMN PID AFTER REPLACEMENT

mean=props["PID"].mean()
props["PID"].replace(np.nan,mean)
```
